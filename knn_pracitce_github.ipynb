```python
import os
train_directory="./train1/"
test_directory="./test2/"

# 테스트와 트레인 데이터 위치 지정 
```

```python
def images(image_directory):
    return [image_directory + image  for image in os.listdir(image_directory)]
# 이미지 디렉토리에 있는 리스트 전부를 뿌려준다. 주소와 이미지 이름을 같이출력
images(train_directory)
```

```python
train_image_names=images(train_directory) #여기서 트레인y 라벨을 뽑아준다.
```

```python
def extract_labels(file_names):
    n=len(file_names)
    y=np.zeros(n,dtype=np.int32)
    
    for i,filename in enumerate(file_names):# 캣에 0을 할당 개에 1할당
        if 'cat'in str(filename):
            y[i]=0
        else:
            y[i]=1
            
    return y

extract_labels(train_image_names)
```

```python
y=extract_labels(train_image_names)
np.save('y',y)
print (y)
```

```python
## 이미지의 사이즈를 얻는 작업
from PIL import Image
image_size = [ ]

for i in train_image_names: # images(file_directory)
    im = Image.open(i)
    image_size.append(im.size) # A list with tuples: [(x, y), …] 
    
#사이즈를 찾아서 이미지 사이즈 리스트에 붙이는 작업

```

```python

STANDARD_SIZE = (28, 28)
def img_to_matrix(filename, verbose=False):# 이미지를 리사이즈하는 작업
        
    #이미지  파일명으로 파일을 불러와서 그걸 rgb픽셀 넘파이로 바꿔주는 작업
    img = Image.open(filename)
    # img = Image.fromarray(filename)
    if verbose == True:
        print ("Changing size from")
        
        # verbose는 여기서 프로그램이 잘돌고 있는지 알려준다. 에러나 예를 보여줌
        # 디버그 관련한 것을 보여주는 false인 경우 안보여 줘도 된다.
    img = img.resize((28,28))
    #이미지 사이즈를 28,28 로 줄여준다.
    
    img = list(img.getdata())
    #픽셀값을 가지는 시퀀스 객체로 만들어준다. 
    img = map(list, img)
    # 리스트로 안에 있는 데이터 값을 맵핑해준다. 캐스팅해준다고 생각하자
    img = np.array(img)
    return img

# Function to flatten numpy array
def flatten_image(img):
    
    '''
    takes in an (m, n) numpy array and flattens it 
    into an array of shape (1, m * n)
    # 이미지를  1차원으로 펴는 작업'''
    s = img.shape[0] * img.shape[1]
    img_wide = img.reshape(1, s)
    return img_wide[0]
```

```python
# 트레이닝 데이타를 준비하자
from PIL import Image
train_data=[]
for i in images(train_directory): #트레인데이터 파일 주소
    img=img_to_matrix(i)
    img=flatten_image(img)
    train_data.append(img)
    
    
print (len(train_data))
```

```python
test_images=images(test_directory)

test = [ ]
for i in test_images:
    img = img_to_matrix(i)
    img = flatten_image(img)
    test.append(img)

test = np.array(test)
test.shape
```
